```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("50_Startups.csv")
X = dataset.iloc[:,0].values
y = dataset.iloc[:,-1].values
```

```python
X = X.reshape(len(y),1)
```

```python
y = y.reshape(len(y),1)
```

```python
plt.scatter(X,y)
plt.title('SVR')
plt.xlabel('R&DSpent')
plt.ylabel('Profit')
plt.show()
```

```python

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

```python
from sklearn.svm import SVR
reg_lin = SVR( kernel = 'linear')
reg_lin.fit(X,y)
```

```python
sc_y.inverse_transform(reg_lin.predict(sc_X.transform([[100000]])).reshape(-1,1))
```

```python
x_grid = np.arange(min(sc_X.inverse_transform(X)), max(sc_X.inverse_transform(X)), 0.1)
x_grid = x_grid.reshape((len(x_grid),1))
plt.scatter(sc_X.inverse_transform(X), sc_y.inverse_transform(y), color = 'grey')
plt.plot(x_grid, sc_y.inverse_transform(reg_lin.predict(sc_X.transform(x_grid)).reshape(-1,1)), color = 'purple')
plt.scatter(100000,134257.98279464, color = 'red')
plt.axvline(x = 100000, ymin = 0, ymax = .67, linestyle = '--')
plt.axhline(y = 134257.98279464, xmax = 0.59, linestyle = '--')
plt.title('SVR')
plt.xlabel('R&DSpent')
plt.ylabel('Profit')
plt.show()
```

```python
error_lin = sum(sc_y.inverse_transform(y) - sc_y.inverse_transform(reg_lin.predict(sc_X.transform(X)).reshape(-1,1)))**2
error_lin
```

```python
reg_lin.score(X, y)
```

```python
from sklearn.svm import SVR
reg_poly = SVR( kernel = 'poly')
reg_poly.fit(X,y)
```

```python
sc_y.inverse_transform(reg_poly.predict(sc_X.transform([[100000]])).reshape(-1,1))
```

```python
x_grid = np.arange(min(sc_X.inverse_transform(X)), max(sc_X.inverse_transform(X)), 0.1)
x_grid = x_grid.reshape((len(x_grid),1))
plt.scatter(sc_X.inverse_transform(X), sc_y.inverse_transform(y), color = 'grey')
plt.plot(x_grid, sc_y.inverse_transform(reg_poly.predict(sc_X.transform(x_grid)).reshape(-1,1)), color = 'purple')
plt.scatter(100000,111017.88333666, color = 'red')
plt.axvline(x = 100000, ymin = 0, ymax = .44, linestyle = '--')
plt.axhline(y = 111017.88333666, xmax = 0.59, linestyle = '--')
plt.title('SVR')
plt.xlabel('R&DSpent')
plt.ylabel('Profit')
plt.show()
```

```python
error_poly = sum(sc_y.inverse_transform(y) - sc_y.inverse_transform(reg_poly.predict(sc_X.transform(X)).reshape(-1,1)))**2
error_poly
```

```python
reg_poly.score(X, y)
```

```python
from sklearn.svm import SVR
reg_rbf = SVR( kernel = 'rbf')
reg_rbf.fit(X,y)
sc_y.inverse_transform(reg_rbf.predict(sc_X.transform([[100000]])).reshape(-1,1))
```

```python
x_grid = np.arange(min(sc_X.inverse_transform(X)), max(sc_X.inverse_transform(X)), 0.1)
x_grid = x_grid.reshape((len(x_grid),1))
plt.scatter(sc_X.inverse_transform(X), sc_y.inverse_transform(y), color = 'grey')
plt.plot(x_grid, sc_y.inverse_transform(reg_rbf.predict(sc_X.transform(x_grid)).reshape(-1,1)), color = 'purple')
plt.scatter(100000,134415.73596733, color = 'red')
plt.axvline(x = 100000, ymin = 0, ymax = .66, linestyle = '--')
plt.axhline(y = 134415.73596733, xmax = 0.59, linestyle = '--')
plt.title('SVR')
plt.xlabel('R&DSpent')
plt.ylabel('Profit')
plt.show()
```

```python
error_rbf = sum(sc_y.inverse_transform(y) - sc_y.inverse_transform(reg_rbf.predict(sc_X.transform(X)).reshape(-1,1)))**2
error_rbf
```

```python
reg_rbf.score(X,y)
```

```python
8.75915229e+12 < 9.48248346e+12 <  1.34847376e+13
```
