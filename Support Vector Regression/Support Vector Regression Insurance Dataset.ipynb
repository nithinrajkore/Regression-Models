```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("insurance.csv")
X = dataset.iloc[:,0].values
y = dataset.iloc[:,-1].values
```

```python
X = X.reshape(len(y),1)
print(X)
```

```python
y = y.reshape(len(y),1)
print(y)
```

```python
plt.scatter(X,y)
plt.title('SVR')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)

print(X)
```

```python
print(y)
```

```python
from sklearn.svm import SVR
reg_rbf = SVR( kernel = 'rbf')
reg_rbf.fit(X,y)
```

```python
sc_y.inverse_transform(reg_rbf.predict(sc_X.transform([[65]])).reshape(-1,1))
```

```python
x_grid = np.arange(min(sc_X.inverse_transform(X)), max(sc_X.inverse_transform(X)), 0.1)
x_grid = x_grid.reshape((len(x_grid),1))
plt.scatter(sc_X.inverse_transform(X), sc_y.inverse_transform(y), color = 'yellow')
plt.plot(x_grid, sc_y.inverse_transform(reg_rbf.predict(sc_X.transform(x_grid)).reshape(-1,1)), color = 'red')
plt.title('SVR')
plt.xlabel('Age') 
plt.ylabel('Charges')
plt.show()
```

```python
error = sum(sc_y.inverse_transform(y) - sc_y.inverse_transform(reg_rbf.predict(sc_X.transform(X)).reshape(-1,1)))**2
error
```

```python
error = sum(sc_y.inverse_transform(y) - sc_y.inverse_transform(reg_linear.predict(sc_X.transform(X)).reshape(-1,1)))**2
error
```
