```python
dataset.isnull().sum()
```

```python
X = dataset.iloc[:,:-1].values
X
```

```python
y = dataset.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print("X_train: ",X_train.shape)
print("X_test: ",X_test.shape)
print("y_train: ",y_train.shape)
print("y_test: ",y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
# Add scaling
```

### Simple Linear Regression

#### 1. Only Ambient Temperature

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train[:,0].reshape(-1,1),y_train)
```

```python
reg.predict([[11.22]])
```

```python
y_pred = reg.predict(X_test[:,0].reshape(-1,1))
```

```python
plt.scatter(X_train[:,0],y_train, color = 'yellow')
plt.plot(X_train[:,0],reg.predict(X_train[:,0].reshape(-1,1)), color = 'green')
plt.xlabel('AT')
plt.ylabel('PE')
plt.title('SIMPLE LINEAR REGRESSION - training set')
```

```python
plt.scatter(X_test[:,0],y_test, color = 'yellow')
plt.plot(X_train[:,0],reg.predict(X_train[:,0].reshape(-1,1)), color = 'green')
plt.xlabel('AT')
plt.ylabel('PE')
plt.title('SIMPLE LINEAR REGRESSION - test set')
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2_score(y_test, y_pred)
```
